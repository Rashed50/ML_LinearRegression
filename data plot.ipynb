```python
import pandas as pd

health_data = {'Duration':[30,30,45,45,45,60,60,60,75,75],
     'Average_Pluse':[80,85,90,95,100,105,110,115,120,125],
     'Max_Pluse':[120,120,130,130,140,140,145,145,150,150],
     'Calorie_Burnage':[240,250,260,270,280,290,300,310,320,330],
     'Hours_Work':[10,10,8,8,0,7,7,8,0,8],
     'Hours_Sleep':[7,7,7,7,7,8,8,8,8,8]
     }
health_data_frame = pd.DataFrame(data=health_data)
#print(health_data_frame)

# count the number of colum
count_col = health_data_frame.shape[1]
#print(count_col)

# count the number of rows
count_row = health_data_frame.shape[0]
#print(count_row)

#We can use the dropna() function to remove the NaNs. axis=0 means that we want to remove all rows tha, Here True means remove otherwise false
health_data_frame.dropna(axis=0,inplace=True)
print('After removing NaNs value \n')
#print(health_data_frame)

import matplotlib.pyplot as plt
# before plot the data we have to clean data. all row data should be formatted and same types we can see data type use info method
print(health_data_frame.info())
# one NaNs value in data so we can change data type. We cannot use objects to calculate and perform analysis here. We must convert the type object to
health_data_frame["Calorie_Burnage"] = health_data_frame['Calorie_Burnage'].astype(int)
health_data_frame.plot(x='Average_Pluse',y='Calorie_Burnage',kind='line')
plt.ylim(ymin=0,)
plt.xlim(xmin=0)
plt.show()

# Import the pyplot module of the matplotlib library
# Plot the data from Average_Pulse against Calorie_Burnage
# kind='line' tells us which type of plot we want. Here, we want to have a straight line
# plt.ylim() and plt.xlim() tells us what value we want the axis to start on. Here, we want the axis to begin from zero
# plt.show() shows us the output

# Slope and Intercept
# Now we will explain how we found the slope and intercept of our function from the linear graph:
# We can find the slope by using the proportional difference of two points from the graph.
# If the average pulse is 80, the calorie burnage is 240
# If the average pulse is 90, the calorie burnage is 260
# f(x) = 2x + 80
# If the average pulse is 80, the calorie burnage is 240
# If the average pulse is 90, the calorie burnage is 260
# We see that if average pulse increases (90-80) with 10, the calorie burnage increases by (260-240) 20.
# Slope = 20/10 = 2
# Mathematically, Slope is Defined as:
# Slope = f(x2) - f(x1) / x2-x1
# f(x2) = Second observation of Calorie_Burnage = 260
# f(x1) = First observation of Calorie_Burnage = 240
# x2 = Second observation of Average_Pulse = 90
# x1 = First observation of Average_Pulse = 80
# Slope = (260-240) / (90 - 80) = 2



```
