```python
# import tkinter   # Google Colab not support tkinter library
# window = tkinter.TK()
# window.title("GUI")
# label = tkinter.Label(window, text="This is Tkinter GUI lirbary ").pack()
# window.mainloop()


import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Sample dataset
data = {
    'Size (sq ft)': [1000, 1500, 2000, 2500, 3000],
    'Bedrooms': [2, 3, 3, 4, 4],
    'Price ($)': [200000, 250000, 280000, 310000, 350000]
}

df = pd.DataFrame(data)

print("\n datafrmae values \n\n ",df)
# Features (X) and target (y)
X = df[['Size (sq ft)', 'Bedrooms']]
y = df['Price ($)']


print("\n datafrmae X values \n\n ",X)

print("\n datafrmae y values \n\n ",y)
# Train Linear Regression Model
model = LinearRegression()
model.fit(X, y)

#print("\n After fit Linear Regression model   \n\n ",model)
# Get coefficients
print("Intercept (b0):", model.intercept_)
print("Coefficients of room_size and bedroom (b1, b2):", model.coef_)

# Define new data for prediction
new_houses = np.array([
    [2500, 4],  # House 1: 2500 sq ft, 4 bedrooms
    [1800, 3],  # House 2: 1800 sq ft, 3 bedrooms
    [3200, 5]   # House 3: 3200 sq ft, 5 bedrooms
])

# Predict prices
predicted_prices = model.intercept_ + np.dot(new_houses, model.coef_)
print(" \nNew House Predicted Prices: \n\n", predicted_prices)

```
