```python
import numpy as np
import sys,time
ls = np.arange(1000) # Uses more memory due to dynamic type
print("memory size of list is %d",sys.getsizeof(5)*len(ls)) # 5 mean integer data type
ar = np.arange(1000)
print("memory size of array is %d",ar.size*ar.itemsize) # array size * every item size
print("dimention , item size , array type, array shape: ",ar.ndim,ar.size,ar.dtype,ar.shape) # findout the array

ar2 = np.array([(8,9,10),(11,12,13)])
print("change array shape 3x3 to 3x2",ar2.reshape(3,2))
print("line space of array ",np.linspace(1,5,10)) #
ar3 = np.array([10,20,30,40,50])
print("mean value and std dev ",np.mean(ar3), ar3.mean(), np.std(ar3) )
print("sqrt of every element",np.sqrt(ar3))

x= np.array([(1,2,3),(3,4,5)])
y= np.array([(1,2,3),(3,4,5)])
print("array subtraction \n",x-y)
print("array multiplication \n",x*y)
print("division \n",x/y)

ar5 = np.array([1,2,3])
print("exponential value of array data \n",np.exp(ar5))
print("log value of array ",np.log(ar5))
print("log base 10 value of array ",np.log10(ar5))


import matplotlib.pyplot as plt
import matplotlib.animation as anm # not use now
x = np.arange(0,np.pi*3,0.1) # value interval
print("\n array data for plot \n",x)
y = np.sin(x)
plt.plot(x,y)
plt.show()  # we can also show using this print(plt.plot(x,y))

y = np.tan(x)
print(plt.plot(x,y)) # tan funtion data plot





```
